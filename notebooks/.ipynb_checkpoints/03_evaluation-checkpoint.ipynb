```python
import os
import sys
sys.path.append(os.path.abspath(os.path.join(os.getcwd(), "..")))
import torch
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.metrics import confusion_matrix, classification_report
from torchvision import transforms
from PIL import Image
```

```python
from src.model import get_model
project_root = os.path.abspath(os.path.join(os.getcwd(), ".."))
model_path = os.path.join(project_root, "outputs", "model.pth")

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = get_model(num_classes=10).to(device)
model.load_state_dict(torch.load(model_path, map_location=device))
model.eval()
```

```python
from src.dataset import get_dataloaders

_, _, test_loader, class_names = get_dataloaders(
    batch_size=64,
    root="data",
    subset_classes=None
)
```

```python
all_preds = []
all_labels = []

with torch.no_grad():
    for images, labels in test_loader:
        images, labels = images.to(device), labels.to(device)
        outputs = model(images)
        preds = torch.argmax(outputs, dim=1)
        all_preds.extend(preds.cpu().numpy())
        all_labels.extend(labels.cpu().numpy())

accuracy = np.mean(np.array(all_preds) == np.array(all_labels))
print(f"Test Accuracy: {accuracy * 100:.2f}%")
```

```python
print(classification_report(all_labels, all_preds, target_names=class_names))
```

```python
cm = confusion_matrix(all_labels, all_preds)
plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt="d", xticklabels=class_names, yticklabels=class_names, cmap="Blues")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

```python
images, labels = next(iter(test_loader))
images, labels = images[:5].to(device), labels[:5].to(device)

with torch.no_grad():
    outputs = model(images)
    probs = torch.nn.functional.softmax(outputs, dim=1)
    top3_probs, top3_preds = torch.topk(probs, 3)

fig, axes = plt.subplots(1, 5, figsize=(20, 5))
for i, ax in enumerate(axes):
    img = transforms.ToPILImage()(images[i].cpu())
    ax.imshow(img)
    ax.axis('off')
    title = "\n".join([f"{class_names[p]} ({top3_probs[i][j]:.2f})" for j, p in enumerate(top3_preds[i])])
    ax.set_title(title, fontsize=9)

plt.tight_layout()
plt.show()
```

```python
# Pick a single image and label from the batch
sample_img = images[0].unsqueeze(0)  # Add batch dimension
sample_label = labels[0].item()

# Get predictions
with torch.no_grad():
    outputs = model(sample_img.to(device))
    probs = torch.nn.functional.softmax(outputs, dim=1)
    top3_probs, top3_idx = torch.topk(probs, k=3)

# Evaluate top prediction
predicted_label = top3_idx[0][0].item()  # Top-1 class
true_label = sample_label

print(f"Ground Truth: {class_names[true_label]}")
print(f"Top Prediction: {class_names[predicted_label]}")

if predicted_label == true_label:
    print("Prediction is correct!")
else:
    print("Prediction is incorrect.")
```

```python
from captum.attr import LayerGradCam, LayerAttribution
from captum.attr import visualization as viz
import matplotlib.pyplot as plt

project_root = os.path.abspath(os.path.join(os.path.dirname(__file__), "..")) if "__file__" in globals() else os.path.abspath("..")
output_dir = os.path.join(project_root, "outputs", "interpretations")
os.makedirs(output_dir, exist_ok=True)

# Init Grad-CAM
gradcam = LayerGradCam(model, model.layer4[-1])

# Visualize for 5 test samples 
images, labels = next(iter(test_loader))
images, labels = images[:5].to(device), labels[:5].to(device)

for i in range(5):
    sample_img = images[i].unsqueeze(0)
    sample_label = labels[i].item()

    # Compute Grad-CAM
    attr = gradcam.attribute(sample_img, target=sample_label)
    upsampled_attr = LayerAttribution.interpolate(attr, sample_img.shape[-2:])

    # Create and save visualization
    fig, _ = viz.visualize_image_attr(
        upsampled_attr[0].cpu().detach().numpy(),
        sample_img.squeeze().permute(1, 2, 0).cpu().numpy(),
        method="heat_map",
        sign="positive",
        show_colorbar=True,
        title=f"Grad-CAM: {class_names[sample_label]}"
    )
    
    fig.savefig(os.path.join(output_dir, f"gradcam_{i}_{class_names[sample_label]}.png"))
    plt.close(fig)
```
